- Bernoulli Naive Bayes is used for binary or boolean features, while Multinomial Naive Bayes is used for discrete features with multiple values.
- Bernoulli Naive Bayes explicitly models the presence or absence of a feature, while Multinomial Naive Bayes models the number of counts of a feature.
- Bernoulli Naive Bayes would be appropriate for classifying spam emails based on the presence or absence of certain words, while Multinomial Naive Bayes would be more appropriate for classifying documents based on the frequency of occurrence of certain words.


Bernoulli Naive Bayes is primarily designed for binary data, where features represent the presence (1) or absence (0) of specific attributes. When dealing with missing values in Bernoulli Naive Bayes

1. Ignoring Missing Values: Bernoulli Naive Bayes typically ignores missing values by treating them as if the attribute is absent (0). This is consistent with the binary nature of the algorithm, where each feature is either present (1) or absent (0). So, if a feature is missing, it's effectively considered as "not present" (0).

2. Imputation: In some cases, you might choose to impute missing values before applying Bernoulli Naive Bayes. Imputation means filling in the missing values with a specific value (either 0 or 1). The choice of imputation depends on your problem and data. For Bernoulli Naive Bayes, you can either impute with 0 (absence) or 1 (presence) based on domain knowledge or the nature of your dataset.

3. Imputing with Probabilities: In more advanced scenarios, you might use statistical techniques to impute missing values probabilistically. For example, you could calculate the probability of an attribute being present (1) based on the overall distribution of that attribute in the dataset and use that probability to impute missing values. This approach attempts to capture the uncertainty associated with imputing missing data.


Yes, Gaussian Naive Bayes can be used for multi-class classification tasks. Gaussian Naive Bayes is an extension of the Naive Bayes algorithm that is specifically designed for continuous data, and it can handle multi-class classification problems where the target variable has more than two classes.

 Gaussian Naive Bayes can handle multi-class classification problems by modeling the data distribution for each class as Gaussian and then calculating the most likely class for a given data point. It's a simple and efficient algorithm for both binary and multi-class classification tasks, although it does make the assumption of feature independence within each class, which may not always hold in practice.

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.3 , random_state=0 )
```

# GUASIAN NAIVE BAYES CLASSIFIER

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(X_train , y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score , precision_score , recall_score , f1_score 
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Calculate precision
precision = precision_score(y_test, y_pred)

# Calculate recall
recall = recall_score(y_test, y_pred)

# Calculate F1 score
f1 = f1_score(y_test, y_pred)
```

```python
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
# cross val = 10 
```

```python
from sklearn.model_selection import cross_val_score
```

```python
gnb1 = GaussianNB()
```

```python
gnb_scores = cross_val_score(gnb1 , X , y , cv = 10)
```

```python
print(np.mean(gnb_scores))
```
